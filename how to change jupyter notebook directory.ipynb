# 怎样修改jupyter notebook的默认路径

上网搜索的步骤是对的，但是总是提示“No such notebook dir: ”，这里需要注意路径的书写方式。修改路径的方法有下面几个步骤：

1、创建目标文件夹，这里修改到D:\jupyter notebook

2、打开“anaconda promt”，运行“jupyter notebook --generate-config”

3、运行“jupyter notebook --generate-config”，在C:\Users\yetian\.jupyter文件夹下找到“jupyter_notebook_config”文件

3、找到“c.NotebookAPP.notebook_dir= ”，并将其改为“c.NotebookApp.notebook_dir = 'D:jupyter notebook'”。

这里需要注意，折腾了很久。我原来写的是'D:\jupyter notebook'但是老报错，说“No such notebook dir: ”,改成'D:jupyter notebook'就好了。

4、打开“anaconda promt”,运行“jupyter notebook”,大功告成。
